```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm 
import statistics 
import seaborn as sns
```

```python
d2 = np.array([145,170,174,168,180,155,162,180,152]).astype(float) 
x_test = np.arange(100,250,1)
prob_x = []

mean = np.mean(d2)
sd = np.std(d2)
prob_x.append(norm.pdf(d2, mean, sd).sum())
sns.lineplot(x=x_test, y=norm.pdf(x_test, mean, sd), label=f'Distribution theta_d2 mean={mean:.2f} sd={sd:.2f}')
plt.axvline(x=mean, color='r', linestyle='--')
mean = 165
sd = 15
prob_x.append(norm.pdf(d2, mean, sd).sum())
sns.lineplot(x=x_test, y=norm.pdf(x_test, mean, sd), label=f'Distribution theta_1 mean={mean:.2f} sd={sd:.2f}')
mean = 166
sd = 10
prob_x.append(norm.pdf(d2, mean, sd).sum())
sns.lineplot(x=x_test, y=norm.pdf(x_test, mean, sd), label=f'Distribution theta_2 mean={mean:.2f} sd={sd:.2f}')
plt.xticks(np.arange(min(x_test),max(x_test),10))
sns.set(rc={'figure.figsize':(10,6)})
plt.xlabel('Value X')
plt.ylabel('P(X)')
plt.legend()
print("Normal Distribution D2 with theta_d2,1,2 (SUM)",prob_x)
plt.show()
```
