```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Generate synthetic data
X, y = make_regression(n_samples=100, n_features=1, noise=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize parameters
lambdas = np.logspace(-4, 4, 100)  # Range of regularization parameters
train_errors = []
test_errors = []

# Train and evaluate for each lambda
for lam in lambdas:
    model = Ridge(alpha=lam)  # Set the regularization parameter
    model.fit(X_train, y_train)
    train_errors.append(np.mean((model.predict(X_train) - y_train) ** 2))  # Mean squared error on training set
    test_errors.append(np.mean((model.predict(X_test) - y_test) ** 2))    # Mean squared error on test set

# Plotting the results
plt.figure(figsize=(12, 6))
plt.semilogx(lambdas, train_errors, label='Training Error', color='blue')
plt.semilogx(lambdas, test_errors, label='Test Error', color='red')
plt.title('Effect of Regularization on Error')
plt.xlabel('Regularization Parameter (lambda)')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.grid(True)
plt.show()

# Training iterations example (simple gradient descent or model fitting)
n_iterations = 100
errors = []
for i in range(n_iterations):
    model = Ridge(alpha=0.1)  # Regularization parameter
    model.fit(X_train, y_train)
    train_errors.append(np.mean((model.predict(X_train) - y_train) ** 2))
    errors.append(model.score(X_train, y_train))  # Store some metric for iterations

# Plot iterations
plt.figure(figsize=(12, 6))
plt.plot(range(n_iterations), errors)
plt.title('Model Performance Over Training Iterations')
plt.xlabel('Iterations')
plt.ylabel('Model Score')
plt.grid(True)
plt.show()
```
